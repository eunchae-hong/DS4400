```python
# Problem 5: Average, variance, and correlation

import csv
import pandas as pd
house_data = pd.read_csv("kc_house_data.csv")
house_data
```

```python
# 1. For each feature, write code to compute the average value, the min and max values, as well as its variance
house_features = house_data.drop(columns = ["id", "date", "zipcode", "price"])
house_price = house_data["price"]
features_stats = house_features.agg(["mean", "min", "max", "var"])
features_stats
```

```python
# Part 1 Answers
high_mean = round(features_stats.loc["mean"].max(), 2)
high_mean_feature = features_stats.loc["mean"].idxmax()
print("The feature with the highest mean is", high_mean_feature, high_mean)

low_mean = round(features_stats.loc["mean"].min(), 2)
low_mean_feature = features_stats.loc["mean"].idxmin()
print("The feature with the lowest mean is", low_mean_feature, low_mean)

high_var = round(features_stats.loc["var"].max(), 2)
high_var_feature = features_stats.loc["var"].idxmax()
print("The feature with the highest variance is", high_var_feature, high_var)

low_var = round(features_stats.loc["var"].min(), 2)
low_var_feature = features_stats.loc["var"].idxmin()
print("The feature with the lowest variance is", low_var_feature, low_var)
```

```python
# 2. Compute the correlation coefficient of each feature with the response. Include a table with the correlation coefficient of each feature with the response. 
# Which features are positively correlated (i.e., have positive correlation coefficient) with the response?
# Which feature has the highest positive correlation with the response?
corr_coeff = house_features.corrwith(house_price)
corr_coeff_table = pd.DataFrame(corr_coeff).reset_index()
corr_coeff_table.columns = ["Feature Name", "Correlation with Price"]
print(corr_coeff_table)

pos_corr = corr_coeff[corr_coeff > 0].index.tolist()
high_corr_coeff = round(corr_coeff.max(), 2)
high_corr_coeff_feature = corr_coeff.idxmax()

print("The features that are positively correlated with the response are", pos_corr)
print("The feature with the highest positive correlation is", high_corr_coeff_feature, high_corr_coeff)

# 3. Were you able to find any features with a negative correlation coefficient with the response?
neg_corr = corr_coeff[corr_coeff < 0].index.tolist()
print("The features that are negatively correlated with the response are", neg_corr)
```
