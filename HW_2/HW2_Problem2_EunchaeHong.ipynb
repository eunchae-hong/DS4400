```python
# DS4400 HW 2
# Eunchae Hong
# Problem 2: Linear Regression
import csv
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import *

house_data = pd.read_csv("kc_house_data.csv")
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
# Problem 2 Part 1
# Report the coefficients of the linear regression models and the following metrics on the training data: (1) MSE metric; (2) R^2 Metric

# train the multiple linear regrssion model
house_features = house_data.drop(columns = ["id", "date", "zipcode", "price"])
house_price = house_data["price"]

X_train = train_data.drop(columns = ["zipcode", "price"])
y_train = train_data["price"]

X_test = test_data.drop(columns = ["id", "date", "zipcode", "price"])
y_test = test_data["price"]

# scale the data using StandardScalar
scale = StandardScaler()
X_train_scaled = scale.fit_transform(X_train)

model_train = LinearRegression()
model_train.fit(X_train_scaled, y_train)

y_train_pred = model_train.predict(X_train_scaled)

# report the coefficients of the linear regression models, the highest coefficient, MSE metric, and R^2 metric
coef = model_train.coef_
print("The coefficients of the linear regression models are:", coef)

feature_names = X_train.columns
max_idx = np.argmax(np.abs(coef)), 2

print("Highest coefficient feature:", feature_names[max_idx])
print("Coefficient value:", coef[max_idx])

train_mse = round(mean_squared_error(y_train, y_train_pred), 2)
print("The MSE metric for the training set is:", train_mse)

train_r_squared = round(r2_score(y_train, y_train_pred), 2)
print("The R Squared Metric for the training set it is:", train_r_squared)
```

```python
# Problem 2 Part 2
# Evaluate the model on the testing set. Report the MSE and R^2 metrics on the testing set.

# scale = StandardScaler()
X_test_scaled = scale.transform(X_test)

y_test_pred = model_train.predict(X_test_scaled)

mse = round(mean_squared_error(y_test, y_test_pred), 2)
print("The MSE metric for the testing set is:", mse)

r_squared = round(r2_score(y_test, y_test_pred), 2)
print("The R Squared Metric for the testing set is:", r_squared)
```
